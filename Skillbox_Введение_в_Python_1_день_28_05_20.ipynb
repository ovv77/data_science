```
# Перезапишем индекс с нумерации строк на даты
data.index = pd.to_datetime(data['Местное время в Москве (ВДНХ)'], dayfirst=True)
```

```
data
```

```
# DataFrame (таблица) позволяет находить данные по условию
# условие записывается в квадратных скобках

last_day = pd.Timestamp(day=1, month=5, year=2016)  # В переменную записали 1 мая 2016
data[data.index < last_day]
```

```
data_cropped = data[data.index < last_day]
plt.plot(data_cropped.index, data_cropped['T'])
```
